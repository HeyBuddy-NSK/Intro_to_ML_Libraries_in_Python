# Introduction to Seaborn

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

## Importing dataset from internet

```python
dataset = sb.load_dataset('diamonds')
print(dataset.head())
```

# Plotting

```python
sb.distplot(dataset['carat'])
plt.show()
```

```python
dataset = sb.load_dataset('tips')
print(dataset.head(10))
```

## Joint Plot

```python
sb.jointplot(x='tip',y='total_bill',data = dataset)
plt.show()
```

```python
dataset = sb.load_dataset('flights')
print(dataset.head(66))
```

## Plotting

```python
sb.catplot(x='month',y = 'passengers', data = dataset, kind = 'violin')
plt.show()
```

```python
dataset = sb.load_dataset('tips')
print(dataset.head(30))
```

```python
graph = sb.FacetGrid(dataset,col = 'sex', hue = 'smoker')
graph.map(plt.scatter, 'total_bill', 'tip')
plt.show()
```

# More codes on seaborn for practice

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing dataset from computer
df = pd.read_csv('Pokemon.csv',index_col = 0, encoding = 'unicode_escape')
df.head(10)
```

## Scatter plot
This function shows regression line with it


```python
sb.lmplot(x = "Attack", y = "Defense", data = df)
plt.show()
```

## Removing Regression Line

```python
sb.lmplot(x='Attack', y='Defense',data = df,fit_reg=False, hue = 'Stage')
plt.show()
```

## Box Plot


```python
df_copy = df.drop(['Total','Stage','Legendary'],axis=1)
print(df_copy.head())

sb.boxplot(data = df_copy)
plt.show()
```

## Violin Plot

```python
sb.violinplot(data = df_copy)
plt.show()
```

```python
# Improving Figure Size
plt.figure(figsize=(10,6))
sb.violinplot(x='Type 1', y = 'Attack', data = df)
plt.show()
```

## Heatmaps

```python
corr = df_copy.corr()    # Correlation of df
sb.heatmap(corr)
```

## Histogram

```python
sb.distplot(df.Attack, color='blue')
plt.show()
```

## Count Plots

```python
sb.countplot(x = 'Type 1', data = df)
plt.xticks(rotation=-45)
```

## Density Plots
Distribution between two attributes

```python
sb.jointplot(df.Attack, df.Defense, kind = 'kde', color = 'lightblue')
plt.show()
```
